### 确定一个编码方式，目前对一条数据的的编码方式为 $ \sum{W_i * Encoder{(feature_i)}}$

```python
weight = [0.1, 0.3, 0.2, 0.2, 0.2]
basic_feature = ['name','cause_of_action','court','type','doc']
test_encoder = []
for index,my_data in test_data_20.iterrows():
    data_encoder = np.zeros(768)
    for i in range(len(weight)):
        data_encoder += weight[i] * model.encode(str(my_data.loc[basic_feature[i]]))
    test_encoder.append( (index,data_encoder) ) # 使用一个二元组存储数据的索引和它的向量编码表示
```

```python
# 注意针对任意一条数据的加权平均编码方式
print(test_encoder[0][1].shape)
```

# encoder写入milvus，由于测试结果不确定性，目前将1000条数据写入milvus

```python
# data_upload
# 总共1494986条数据，选择1000条数据进行编码测试
# test_data = df_test.sample(n=1000)
from milvus import Milvus, IndexType, MetricType, Status
__HOST = '127.0.0.1'
__PORT = '19530'
# __PORT = '19121'

__DIM = 768
__INDEX_FILE_SIZE = 1024 # max file stored


def milvus_build(name, metric_type=None):
    """
    input parameter:
    name: Wanting Create Collection Name, data vector space name
    metric_type: way to compute similiarity
    output parameter:
    instance of our milvus server
    """
    milvus = Milvus(__HOST, __PORT)

    collection_name = name

    # create a vector space named collection_name
    status, ok = milvus.has_collection(collection_name)
    if not ok:
        param = {
                'collection_name': collection_name,
                'dimension': __DIM,
                'index_file_size': __INDEX_FILE_SIZE,  # optional
                'metric_type': MetricType.L2 if metric_type == None else metric_type  # optional
        }
    
        milvus.create_collection(param)
        print("Create "+name+" Successfully")

    _, collections = milvus.list_collections()
    print("Now our collections list: \n", [content + '\n' for content in collections])

    return milvus
```

```python
weight = [0.1, 0.3, 0.2, 0.2, 0.2]
basic_feature = ['name','cause_of_action','court','type','doc']
data_1000_encoder = []
for index,my_data in test_data.iterrows():
    data_encoder = np.zeros(768)
    for i in range(len(weight)):
        data_encoder += weight[i] * model.encode(str(my_data.loc[basic_feature[i]]))
    data_1000_encoder.append( (index,data_encoder) ) # 使用一个二元组存储数据的索引和它的向量编码表示
```

```python
vectors_milvus = []
for data in data_1000_encoder:
    vectors_milvus.append(list(data[1]))
print(len(vectors_milvus))
```

```python
my_milvus = milvus_build('New_Test_Data')
```

# 将向量编码存入Milvus向量索引空间，获取向量相似度检索相相关的索引编码

```python
status, ids = my_milvus.insert(collection_name='New_Test_Data', records= vectors_milvus )
if not status.OK():
        print("Insert failed: ".format(status))
```

```python
print(ids)
```

```python
# ids是Milvus进行向量化搜索返回的索引值，该索引值可以和之前的存储索引构成缓存散列表
milvus_hash_map = {}
for i in range(len(data_1000_encoder)):
    milvus_hash_map[ids[i]] = data_1000_encoder[i][0]
```

```python
# 哈希索引举例,将milvus的hash索引和数据表的hash编码关联
print(milvus_hash_map[1670308218326938001])
```

```python
# 将milvus_hash_map结果存放到json中，该向量编码可以以后使用
import json

with open('test_data_hash.json', 'w') as f:
    json.dump(milvus_hash_map, f)
```
